# 데이터 수집 - 크롤링

### 소상공인시장진흥공단에서 제공하는 상권정보 크롤링
- https://sg.sbiz.or.kr/godo/stat/area.sg

#### 라이브러리

```python
import pandas as pd
```

#### selenium(XPATH)으로 url에 접속

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver = webdriver.Chrome('./driver/chromedriver.exe')
url = 'https://sg.sbiz.or.kr/godo/stat/area.sg'
driver.get(url)
```

```python
# 군산, 목포, 통영 페이지 순서대로 들어가서 접근

# //*[@id="tableBody"]/tr[1]/td[1]  => 행정동
# //*[@id="tableBody"]/tr[1]/td[3]  => 인구수
# //*[@id="tableBody"]/tr[1]/td[7]  => 업소수
```

#### 군산 페이지 정보 가져오기 (dong:행정지역, pop:인구수, num_shop:업소수)

```python
dong = []
pop = []
num_shop = []

for i in range(1, len(driver.find_elements(By.CSS_SELECTOR, '#tableBody tr'))+1) :
    dong.append(driver.find_elements(By.XPATH, f'//*[@id="tableBody"]/tr[{i}]/td[1]')[0].text)
    pop.append(driver.find_elements(By.XPATH, f'//*[@id="tableBody"]/tr[{i}]/td[3]')[0].text)
    num_shop.append(driver.find_elements(By.XPATH, f'//*[@id="tableBody"]/tr[{i}]/td[7]')[0].text)

print('dong list : ', dong)
print('pop list : ', pop)
print('num_shop : ', num_shop)
```

#### 가져온 정보(list)로 데이터프레임(gunsan_df)만들기 

```python
gunsan_df = pd.DataFrame(data = list(zip(dong, pop, num_shop)), columns = ['dong', 'population', 'num_of_shop'])
gunsan_df
```

```python
gunsan_df['dong'] = gunsan_df['dong'].str.replace('┗', '')
```

```python
gunsan_df
```

#### 수집 데이터 저장

```python
gunsan_df.to_excel('./data/gunsan.xlsx', index=False)
```
