```python
# 데이터 불러오기
import numpy as np
import pandas as pd

tmp_frm = pd.read_excel('samsung.xlsx')
```

```python
# 서브셋 만들기
subset_frm = tmp_frm[['기준년월', '법정동명', '업종명_중분류', 'CNT']]
```

```python
# 법정동명 리스트에 담기
lst = subset_frm['법정동명'].str.split()

# 통영시 데이터 인덱스 추출
idx_lst = []
for idx in range(len(lst)):
    if lst[idx][1] == '통영시':
        idx_lst.append(idx)
        
# 통영 데이터프레임 만들기
tongyeong_frm = subset_frm.iloc[idx_lst]
print(tongyeong_frm.head())
```

```python
# 인덱스 리셋
tongyeong_frm.reset_index(inplace=True)
```

```python
# 드랍
tongyeong_frm.drop('index', axis=1, inplace = True)
```

```python
# 시/동 구분
si_lst = []
dong_lst = []
for idx in range(len(tongyeong_frm['법정동명'].str.split())):
    si_lst.append(tongyeong_frm['법정동명'].str.split()[idx][1])
    dong_lst.append(tongyeong_frm['법정동명'].str.split()[idx][2])
```

```python
tongyeong_frm['Si'] = si_lst
tongyeong_frm['Dong'] = dong_lst
tongyeong_frm
```

```python
tongyeong_frm.drop('법정동명', axis=1, inplace = True)
```

```python
tongyeong_frm.rename(columns = {'기준년월' : 'Date', '업종명_중분류':'Shop'}, inplace = True)
```
