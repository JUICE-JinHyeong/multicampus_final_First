#### 파이썬에서 엑셀 가져오기
-(삼성카드)가맹점 기반 지역별 소비행태.xlsx 파일 불러오기  
-군산시에 관한 자료  
-기준년월 = Date, 법정동명 = area, 업종명_중뷴류 = Shop으로 컬럼명 변경  
-군산시 / 개복동 의 형태로 제작
-엑셀의 A : Y 행 사용. 단 B와 D열 제거

```python
import pandas as pd
```

```python
#gs = 군산
#엑셀의 A ~ Y행까지 추출
gs_df = pd.read_excel('C:/Users/ohans/Desktop/multicampus/big data/class/project/final project/data/(삼성카드)가맹점 기반 지역별 소비행태.xlsx',
                     usecols = 'A:Y')
```

```python
gs_df
```

원하는 자료만 추출

```python
#B, D열 제거
gs_df2 = gs_df.drop(['법정동코드','업종코드_중분류'],axis=1)
```

```python
gs_df2.head()
```

```python
#컬럼명 변경
gs_df3 = gs_df2.rename(columns={'기준년월' :'Date',
                                '법정동명' :'area',
                                 '업종명_중분류' :'Shop'})
```

```python
gs_df3.head()
```

```python
#군산시 추출
gs_df4 = gs_df3[gs_df3['area'].str.contains('군산시', na = False)]
```

```python
gs_df4
```

```python
#중복되는 '전북' 제거
gs_df4['area']=gs_df4['area'].str.replace('전북 ','')
```

```python
gs_df4.head()
```

```python
#area의 군산시와 지역명을 분리
#expand = True를 할 경우 split값이 바로 데이터프레임으로 추출
gs_area = gs_df4['area'].str.split(' ',expand = True)
```

```python
gs_area.head()
```

```python
gs_area.columns =['Si','Dong']
```

```python
gs_area
```

```python
#분리한 SI, Dong컬럼을 원 컬럼과 결합
gs_df5 = pd.concat([gs_df4, gs_area], axis = 1)
```

```python
gs_df5.head()
```

```python
#area 컬럼 제거
gs_df6 = gs_df5.drop(['area'],axis=1)
```

```python
gs_df6.head()
```

```python
#Si와 Dong 멀티인덱스
gs_df7 = gs_df6.set_index(["Si","Dong"])
```

```python
gs_df7.head()
```

```python
#보기 좋게 Dong별로 분류
gs_df8 = gs_df7.sort_values('Dong', ascending = True)
```

```python
gs_df8.head()
```
