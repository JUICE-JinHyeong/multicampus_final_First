```python
import pandas as pd
import numpy as np

# 삼성카드 엑셀 파일의 ACEF 칼럼을 ['Date' , 'Si' , 'Shop' , 'cnt'] 로 이름지어 불러옵니다.
# 성 연령별 데이터 생성
# 일별 및 평일, 주말 데이터 생성

data = pd.read_excel('C:/Users/a/Downloads/연습프로젝트/삼성카드 주제/(삼성카드)가맹점 기반 지역별 소비행태.xlsx'
             , usecols = "A,C,E,F" , names = ['Date' , 'Si' , 'Shop' , 'cnt'])
gender_age = pd.read_excel('C:/Users/a/Downloads/연습프로젝트/삼성카드 주제/(삼성카드)가맹점 기반 지역별 소비행태.xlsx'
             , usecols = "G:P")
day_of_weekend = pd.read_excel('C:/Users/a/Downloads/연습프로젝트/삼성카드 주제/(삼성카드)가맹점 기반 지역별 소비행태.xlsx'
             , usecols = "Q:Y")
```

```python

# 목포시 데이터만을 남기기 위해 목포시 단어를 포함하고 있는 행을 찾아 추출하고 index 값을 저장합니다.

mokpo_idx = []
for indx , value in enumerate(data.Si) :
    if '목포시' in value :
        print(value)
        mokpo_idx.append(indx)
    
# 저장된 index 값으로 원래 데이터 data에 넣고 인덱스를 리셋하여 재정렬해줍니다.
# 목포시가 담긴 데이터가 만들어졌습니다.

mokpo_frm = data.loc[mokpo_idx].reset_index(drop=True)
#mokpo_frm.head()
        
# 성연령별 데이터 프레임
gender_age_frm = gender_age.loc[mokpo_idx].reset_index(drop=True)
#gender_age_frm.head()

# 일별 및 평일, 주말 데이터 생성
day_of_weekend_frm = day_of_weekend.loc[mokpo_idx].reset_index(drop=True)
# day_of_weekend_frm.head()

# 시 와 동 으로 분류하기 위해 split(' ')으로 단어를 분류하고 2번째에 들어있는 도로명 , 동 , 읍 , 면 리 등의 값을 가져옵니다.

dong_lst = []
for i in range(len(mokpo_frm)) :
    dong_lst.append(mokpo_frm['Si'][i].split(' ')[2])
    
    
# 시 와 동 으로 분류하기 위해 split(' ')으로 단어를 분류하고 2번째에 들어있는 도로명 , 동 , 읍 , 면 리 등의 값을 가져옵니다.

dong_lst = []
for i in range(len(mokpo_frm)) :
    dong_lst.append(mokpo_frm['Si'][i].split(' ')[2])

# 시와 동 칼럼을 새로 만들되, 동 칼럼은 시 칼럼 뒤에 있어야 하므로 insert 함수를 사용합니다.

mokpo_frm.insert(2 , 'Dong' , dong_lst)
mokpo_frm['Si']   = '목포시'
```

```python
# concat으로 원하는 데이터 만들기

# 1. 목포시 성연령별 데이터

mokpo_gender_age_frm = pd.concat([mokpo_frm , gender_age_frm] , axis = 1)
display(mokpo_gender_age_frm.head())


# 2. 목포시 요일별 데이터

mokpo_day_of_weekend_frm = pd.concat([mokpo_frm , day_of_weekend_frm] , axis = 1)
display(mokpo_day_of_weekend_frm.head())
```
