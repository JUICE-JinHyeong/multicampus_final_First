```python
import numpy as np
from time import sleep
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# 코카콜라 제로 250ml 30캔
driver = webdriver.Chrome('../driver/chromedriver.exe')
url = 'https://smartstore.naver.com/cocakorea/products/708741692?n_media=11068&n_query=%EC%BD%94%EC%B9%B4%EC%BD%9C%EB%9D%BC%EC%A0%9C%EB%A1%9C&n_rank=2&n_ad_group=grp-a001-02-000000027710026&n_ad=nad-a001-02-000000188925969&n_campaign_type=2&n_mall_id=cocakorea&n_mall_pid=708741692&n_ad_group_type=2&NaPm=ct%3Dlc8q19xk%7Cci%3D0A80001zbUPxpGjEVuZq%7Ctr%3Dpla%7Chk%3Dfd3118f614fd30357547ce380ba35cb7d12a4f2e'
driver.get(url)
```

```python
driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[1]/div/div[1]/div/div[1]/div/div/div[2]/div/span')[0].text
```

```python
review_lst = driver.find_elements(By.XPATH, '//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li')
len(review_lst)
```

```python
lst_250ml_30 = []
for i in np.arange(8,13):
    for idx in range(len(review_lst)):
        data = driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{idx+1}]/div/div[1]/div/div[1]/div/div/div[2]/div/span')[0].text
        if data == '한달사용기' or data == '재구매':
            data = driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{idx+1}]/div/div[1]/div/div[1]/div/div/div[2]/div/span[2]')[0].text
            if data == '한달사용기' or data == '재구매':
                data = driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{idx+1}]/div/div/div/div[1]/div/div/div[2]/div/span[3]')[0].text
            lst_250ml_30.append(data)
        else :
            lst_250ml_30.append(data)
        
    driver.find_element(By.XPATH , f'//*[@id="REVIEW"]/div/div[3]/div[2]/div/div/a[{i}]').click()
    sleep(3)    
  
```
