```python
import numpy as np
from time import sleep
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# 코카콜라 제로 355ml
```

```python
driver = webdriver.Chrome('../driver/chromedriver.exe')
url = 'https://search.shopping.naver.com/catalog/27817802526?query=%EC%BD%94%EC%B9%B4%EC%BD%9C%EB%9D%BC%20%EC%A0%9C%EB%A1%9C&NaPm=ct%3Dlc8e11qw%7Cci%3D0cd43af612e1f6346a394c8dc228d0f00decc38d%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3Db5deced91dab05ff383cbb48de441ea8b555742b'
driver.get(url)
```

```python
driver.find_elements(By.XPATH, f'//*[@id="section_review"]/ul/li[11]/div[2]/div[1]/p')[0].text
```

```python
review_lst = driver.find_elements(By.XPATH, '//*[@id="section_review"]/ul/li')
len(review_lst)
```

```python
lst = []
for i in np.arange(2,13):
    for idx in range(len(review_lst)):
        lst.append(driver.find_elements(By.XPATH, f'//*[@id="section_review"]/ul/li[{idx+1}]/div[2]/div[1]/p')[0].text)
        
    driver.find_element(By.XPATH , f'//*[@id="section_review"]/div[3]/a[{i}]').click()
    sleep(3)    


for page in range(10):
    for i in np.arange(3,13):
        for idx in range(len(review_lst)):
            lst.append(driver.find_elements(By.XPATH, f'//*[@id="section_review"]/ul/li[{idx+1}]/div[2]/div[1]/p')[0].text)

        driver.find_element(By.XPATH , f'//*[@id="section_review"]/div[3]/a[{i}]').click()
        sleep(3)    
    
```
