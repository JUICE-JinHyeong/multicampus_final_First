```python
import numpy as np
from time import sleep
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# 부르르 사이다 1.5L 12페트
driver = webdriver.Chrome('../driver/chromedriver.exe')
url = 'https://brand.naver.com/ilhwa/products/5609471720?n_media=11068&n_query=%EB%B6%80%EB%A5%B4%EB%A5%B4%EC%82%AC%EC%9D%B4%EB%8B%A4&n_rank=1&n_ad_group=grp-a001-02-000000021513659&n_ad=nad-a001-02-000000135869737&n_campaign_type=2&n_mall_id=ilhwa01004&n_mall_pid=5609471720&n_ad_group_type=2&NaPm=ct%3Dlca5k1jc%7Cci%3D0A40001HwoTxGFEXb0Z4%7Ctr%3Dpla%7Chk%3D0387559f50dc526e4279fd9145cf351bbad084d4'
driver.get(url)
```

```python
driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[9]/div/div/div/div[1]/div/div[1]/div[2]/div/span[3]')[0].text
```

```python
review_lst = driver.find_elements(By.XPATH, '//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li')
len(review_lst)
```

```python
lst_1500ml_12 = []

for i in np.arange(2,13):
    for idx in range(len(review_lst)):
        data = driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{idx+1}]/div/div[1]/div/div[1]/div/div/div[2]/div/span')[0].text
        if data == '한달사용기' or data == '재구매' or data == 'BEST' :
            data = driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{idx+1}]/div/div[1]/div/div[1]/div/div/div[2]/div/span[2]')[0].text
            if data == '한달사용기' or data == '재구매' or data == 'BEST':
                data = driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{idx+1}]/div/div/div/div[1]/div/div/div[2]/div/span[3]')[0].text
                if data == '한달사용기' or data == '재구매' or data == 'BEST':
                    data = driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{idx+1}]/div/div/div/div[1]/div/div[1]/div[2]/div/span[3]')[0].text
                
            lst_1500ml_12.append(data)
        else :
            lst_1500ml_12.append(data)  
    driver.find_element(By.XPATH , f'//*[@id="REVIEW"]/div/div[3]/div[2]/div/div/a[{i}]').click()
    sleep(3)    
```
