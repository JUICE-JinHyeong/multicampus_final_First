- By.ID	태그의 id값으로 추출
- By.NAME	태그의 name값으로 추출
- By.XPATH	태그의 경로로 추출
- By.LINK_TEXT	링크 텍스트값으로 추출
- By.PARTIAL_LINK_TEXT	링크 텍스트의 자식 텍스트 값을 추출
- By.TAG_NAME	태그 이름으로 추출
- By.CLASS_NAME	태그의 클래스명으로 추출
- By.CSS_SELECTOR	css선택자로 추출

```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url = 'https://prod.danawa.com/info/?pcode=3312370&keyword=%ED%83%84%EC%82%B0%EC%88%98&cate=1623180'

driver = webdriver.Chrome('diver/chromedriver.exe')
driver.get(url)
```

```python
def review_crawling_data(lst : list) :
    for i in range(1, len(driver.find_elements(By.CLASS_NAME , 'atc'))+1):
        lst.append(driver.find_element(By.XPATH , f'/html/body/div[2]/div[3]/div[2]/div[4]/div[4]/div/div[3]/div[2]/div[3]/div[2]/div[5]/ul/li[{i}]/div[2]/div/div[2]').text)
```

```python
# 대기 시간
wait_time = 2
result = []
def crawling_page_load_and_input_review_data(count : int) :
    # 로딩 완료 확인
    WebDriverWait(driver, wait_time).until(
                    EC.presence_of_element_located(
                        (By.XPATH,'//*[@id="product_detail"]')
            )
        )
    count_page = 0
    count_page = count_page + (count * 10)
    for page in range(1, 11) :
        WebDriverWait(driver, wait_time).until(
                    EC.visibility_of_all_elements_located(
                        (By.CLASS_NAME,'danawa-prodBlog-companyReview-clazz-more')
            )
        )
        # 리뷰 한페이지 리스트에 넣기
        count_page = count_page +1
        print(count_page)
    
        time.sleep(1)

        # 크롤링
        review_crawling_data(result)

        if page == 10 :
            if count_page == 10 :
                driver.find_element(By.XPATH 
                                , '/html/body/div[2]/div[3]/div[2]/div[4]/div[4]/div/div[3]/div[2]/div[3]/div[2]/div[5]/div/div/a/span').click()
            else :
                driver.find_element(By.XPATH 
                                , '/html/body/div[2]/div[3]/div[2]/div[4]/div[4]/div/div[3]/div[2]/div[3]/div[2]/div[5]/div/div/a[2]/span').click()
        else :
            driver.find_element(By.XPATH 
                                , f'/html/body/div[2]/div[3]/div[2]/div[4]/div[4]/div/div[3]/div[2]/div[3]/div[2]/div[5]/div/div/div/a[{page}]').click()
    
        
```

```python
# 1 ~ 944 page
# 중간에 멈춤

for repeat in range(1000) :
    crawling_page_load_and_input_review_data(repeat)
```
