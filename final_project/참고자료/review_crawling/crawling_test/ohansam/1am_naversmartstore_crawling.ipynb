```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
# 추가 패키지
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 비동기 함수
# 참고 : https://kukuta.tistory.com/345
import asyncio
```

```python
wait_time = 1
result = []

url = 'https://smartstore.naver.com/pickydiet/products/5135269051?'

driver = webdriver.Chrome('C:/Users/ohans/driver/chromedriver_win32/chromedriver.exe')
driver.get(url)
```

```python
# 리뷰 크롤링(한페이지) 함수 생성

def review_data(lst : list) :
    for i in range(1,len(driver.find_elements(By.CSS_SELECTOR , '._2389dRohZq'))+1) :
        # XPATH 주소 확인(맨 끝)
        # data1 = span[1] or span
        # data2 = span[2]
        data1 = driver.find_element(By.XPATH , f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{i}]/div/div/div/div[1]/div/div[1]/div[2]/div/span').text
        if data1 == '재구매' or data1 == '한달사용기':
            # 첫 번째 if을 통과해야만 data2가 생성이 가능하다.
            data2 = driver.find_element(By.XPATH , f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{i}]/div/div/div/div[1]/div/div[1]/div[2]/div/span[2]').text
            if data2 == '재구매' :
                lst.append(driver.find_element(By.XPATH , f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{i}]/div/div/div/div[1]/div/div[1]/div[2]/div/span[3]').text)
            else :
                lst.append(driver.find_element(By.XPATH , f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{i}]/div/div/div/div[1]/div/div[1]/div[2]/div/span[2]').text)
        else :
            lst.append(data1)
```

```python

def crawling_page_load_and_input_review_data() :
    # 로딩 완료 확인
    WebDriverWait(driver, wait_time).until(
                    EC.presence_of_element_located(
                        (By.XPATH,'//*[@id="REVIEW"]/div/div[3]/div[2]')
            )
        )
    
    WebDriverWait(driver, wait_time).until(
                EC.visibility_of_all_elements_located(
                    (By.CLASS_NAME,'_2389dRohZq')
        )
    )
    # 리뷰 한페이지 리스트에 넣기
    
    review_data(result)
    time.sleep(10)
    driver.implicitly_wait(10)
    driver.find_elements(By.CLASS_NAME , 'fAUKm1ewwo._2Ar8-aEUTq')[0].send_keys(Keys.ENTER)
```

```python
from selenium.webdriver.common.keys import Keys
driver.find_elements(By.CLASS_NAME , 'fAUKm1ewwo._2Ar8-aEUTq')[0].send_keys(Keys.ENTER)
```

```python
# list에 5천개 이상 생기면 너무 무거워진다.
# 5천개씩 끊자.
# 방법 : wait time 추가 혹은 구글 드라이버 재실행 혹은 result 리스트 초기화
for i in range(10001) :
    crawling_page_load_and_input_review_data()
```
