- By.ID	태그의 id값으로 추출
- By.NAME	태그의 name값으로 추출
- By.XPATH	태그의 경로로 추출
- By.LINK_TEXT	링크 텍스트값으로 추출
- By.PARTIAL_LINK_TEXT	링크 텍스트의 자식 텍스트 값을 추출
- By.TAG_NAME	태그 이름으로 추출
- By.CLASS_NAME	태그의 클래스명으로 추출
- By.CSS_SELECTOR	css선택자로 추출

nodename	노드명이 "nodename"인 노드 선택

..	현재 노드의 부모노드 선택
@	현재 노드의 속성 선택
