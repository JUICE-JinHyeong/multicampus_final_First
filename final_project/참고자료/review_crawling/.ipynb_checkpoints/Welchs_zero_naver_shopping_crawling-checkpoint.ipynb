- By.ID	태그의 id값으로 추출
- By.NAME	태그의 name값으로 추출
- By.XPATH	태그의 경로로 추출
- By.LINK_TEXT	링크 텍스트값으로 추출
- By.PARTIAL_LINK_TEXT	링크 텍스트의 자식 텍스트 값을 추출
- By.TAG_NAME	태그 이름으로 추출
- By.CLASS_NAME	태그의 클래스명으로 추출
- By.CSS_SELECTOR	css선택자로 추출

xpath 문법

표현	설명
nodename	노드명이 "nodename"인 노드 선택
/	루트 노드로부터 선택
//	현재 노드로부터 문서상의 모든 노드를 조회
.	현재 노드 선택
..	현재 노드의 부모노드 선택
@	현재 노드의 속성 선택
 

```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url = 'https://search.shopping.naver.com/catalog/32917077620?query=%EC%9B%B0%EC%B9%98%EC%8A%A4%20%EC%A0%9C%EB%A1%9C&NaPm=ct%3Dlc8etu0g%7Cci%3D33fa0630dd673a45193ea111bcffd2a489d01b11%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3D08faec4ceaef1707aa616df5dd22c14fded2055e'

driver = webdriver.Chrome('diver/chromedriver.exe')
driver.get(url)
```

```python
# 리뷰 길이 재기
# reviewItems_text__XrSSf
len(driver.find_elements(By.CLASS_NAME , 'reviewItems_text__XrSSf'))
# 20

def review_crawling_data(lst : list) :
    for i in range(1, len(driver.find_elements(By.CLASS_NAME , 'reviewItems_text__XrSSf'))+1) :
        lst.append(driver.find_element(By.XPATH , f'//*[@id="section_review"]/ul/li[{i}]/div[2]/div[1]/p').text)
```

```python
# 네이버 쇼핑 11 페이지 부터 크롤링시 이 함수 사용

# 대기 시간
wait_time = 2
result = []
def crawling_page_load_and_input_review_data() :
    # 로딩 완료 확인
    WebDriverWait(driver, wait_time).until(
                    EC.presence_of_element_located(
                        (By.XPATH,'//*[@id="section_review"]')
            )
        )
    for page in range(2, 12) :
        WebDriverWait(driver, wait_time).until(
                    EC.visibility_of_all_elements_located(
                        (By.CLASS_NAME,'reviewItems_review__DqLYb')
            )
        )
        # 리뷰 한페이지 리스트에 넣기
        review_crawling_data(result)
        time.sleep(1)
        driver.find_elements(By.XPATH , '//*[@id="section_review"]/div[3]/a')[page].click()
```

```python
# 네이버 쇼핑 1 ~ 10 페이지 크롤링시 이 함수 사용

# 대기 시간
wait_time = 2
result = []
def crawling_page_short() :
    # 로딩 완료 확인
    WebDriverWait(driver, wait_time).until(
                    EC.presence_of_element_located(
                        (By.XPATH,'//*[@id="section_review"]')
            )
        )
    for page in range(1, 11) :
        WebDriverWait(driver, wait_time).until(
                    EC.visibility_of_all_elements_located(
                        (By.CLASS_NAME,'reviewItems_review__DqLYb')
            )
        )
        # 리뷰 한페이지 리스트에 넣기
        review_crawling_data(result)
        time.sleep(1)
        driver.find_elements(By.XPATH , '//*[@id="section_review"]/div[3]/a')[page].click()
```

```python
# 모든 값은 result(list)에 저장된다.
```

```python
# 1 ~ 10
crawling_page_short()
```

```python
# 11 ~ 110
for x in range(100) :
    crawling_page_load_and_input_review_data()
```
