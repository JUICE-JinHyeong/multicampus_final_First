```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://search.shopping.naver.com/catalog/26120149527?query=%EC%9B%90%EC%97%90%EC%9D%B4%EC%97%A0%EC%8A%A4%ED%8C%8C%ED%81%B4%EB%A7%81"
result = requests.get(url)

bs_obj = BeautifulSoup(result.content, "html.parser")

ul=bs_obj.find("ul",{"class":"reviewItems_list_review__q726A"})
print(ul)
```

```python
boxes = ul.findAll("div","reviewItems_review__DqLYb")

for box in boxes:
    print(box)
```

```python
boxes = ul.findAll("div","reviewItems_review__DqLYb")


#1페이지 크롤링
for box in boxes:
    ptag = box.find("p",{"class":"reviewItems_text__XrSSf"})
    print(ptag)
```
